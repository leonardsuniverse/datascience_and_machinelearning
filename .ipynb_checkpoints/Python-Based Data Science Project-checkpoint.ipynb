# Analysing GDP: Use Case

```python
# Gross Domestic Product (GDP) is the final value of all the economic goods
# and services produced within the country's geographic boundaries during a 
# specified period of time. GDP growth rate is the major indicator of a
# country's economic performance. Broadly speaking, the primary sector
# (agricultural), the secondary sector (industry), and the tertiary sector
# (services) all contribute to GDP by producing goods and services
# (services). Gross Domestic Product (GDP) is a key tool that guides
# investors, policymakers,and businesses in strategic decision-making. Per
# capita GDP is a global indicator of a country's economy that economist
# use in combination with GDP to assess a country's wealth based on its
# economic growth. The formula of GDP per capita is:
# GDP per capita = Gross Domestic Product (GDP) / Population
```

# Data Preparation

```python
# The missing data in table is filled in by using the median of the region to
# which a country belongs because geologically close countries are often
# similar in many aspects.
```

```python
import pandas as pd
data = pd.read_csv('world.csv',decimal=',')
medians = data.groupby('Region')[['GDP ($ per capita)','Literacy (%)','Agriculture']].median()
for col in data.columns:
    if data[col].isnull().sum() == 0:
        continue
    if col == 'climate':
        guess_values = data.groupby('Region')['Climate'].apply(lambda x: mode(x).max(0))
    else:
        guess_values = data.groupby('Region')[col].median()
    for region in data['Region'].unique():
        data[col].loc[(data[col].isnull()) & (data['Region']==region)] = guess_values[region]
```

# Top Factors affecting GDP per capita

```python
# The columns that are mostly correlated to GDP per capita were picked and
# scatter plots were created
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20,12))
plt.subplots_adjust(hspace=0.4)

corr_to_gdp = pd.Series(dtype='float64')
for col in data.columns.values[2:]:
    if ((col!='GDP ($ per capita)')&(col!='Climate')):
        corr_to_gdp[col] = data['GDP ($ per capita)'].corr(data[col])
abs_corr_to_gdp = corr_to_gdp.abs().sort_values(ascending=False)
corr_to_gdp = corr_to_gdp.loc[abs_corr_to_gdp.index]

for i in range(2):
    for j in range(3):
        sns.regplot(x=corr_to_gdp.index.values[i*3+j], y='GDP ($ per capita)', data=data,
                   ax=axes[i,j], fit_reg=False, marker='.')
        title = 'correlation='+str(corr_to_gdp[i*3+j])
        axes[i,j].set_title(title)
axes[1,2].set_xlim(0,102)
plt.show()
```

# Modeling

# Training and Testing

```python
# All the features given in the data set are used without further feature
# engineering
```

```python
LE = LabelEncoder()
data['Region_label'] = LE.fit_transform(data['Region'])
data['Climate_label'] = LE.fit_transform(data['Climate'])
data.head()
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
data = pd.read_csv('world.csv',decimal=',')
train, test = train_test_split(data, test_size=0.3, shuffle=True)
training_features = ['Population', 'Area (sq. mi.)',
       'Pop. Density (per sq. mi.)', 'Coastline (coast/area ratio)',
       'Net migration', 'Infant mortality (per 1000 births)',
       'Literacy (%)', 'Phones (per 1000)',
       'Arable (%)', 'Crops (%)', 'Other (%)', 'Birthrate',
       'Deathrate', 'Agriculture', 'Industry']
target = 'GDP ($ per capita)'
# Extract features and target variable from the data
X = data[training_features]
Y = data[target]
# Split data into training and testing sets
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.3, random_state=42)
train_X = train[training_features]
train_Y = train[target]
test_X = test[training_features]
test_Y = test[target]
```

# First, the linear regression model is used. Both root mean squared error and were assessed, as for metric.

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_squared_log_error
import pandas as pd
import numpy as np

# Drop rows with NaN values in the target variable
data.dropna(subset=[target], inplace=True)

# Extract features and target variable from the data
X = data[training_features]
Y = data[target]
# Split data into training and testing sets
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.3, random_state=42)

# Assuming train_X, train_Y, test_X, test_Y are defined and contain your training and testing data

# Instantiate a Linear Regression model
model = LinearRegression()

# Instantiate a SimpleImputer to handle missing values
imputer = SimpleImputer(strategy='most_frequent')

# Fit the imputer on the training data and transform both training and testing data
train_X_imputed = imputer.fit_transform(train_X)
test_X_imputed = imputer.transform(test_X)

# Fit the model on the imputed training data
model.fit(train_X_imputed, train_Y)

# Predict on the imputed training and testing data
train_pred_Y = model.predict(train_X_imputed)
test_pred_Y = model.predict(test_X_imputed)

# Clip negative predictions to 0
train_pred_Y = pd.Series(train_pred_Y.clip(0, train_pred_Y.max()), index=train_Y.index)
test_pred_Y = pd.Series(test_pred_Y.clip(0, test_pred_Y.max()), index=test_Y.index)

# Calculate RMSE and MSLE for both training and testing data
rmse_train = np.sqrt(mean_squared_error(train_pred_Y, train_Y))
msle_train = mean_squared_log_error(train_pred_Y, train_Y)
rmse_test = np.sqrt(mean_squared_error(test_pred_Y, test_Y))
msle_test = mean_squared_log_error(test_pred_Y, test_Y)

# Print evaluation metrics
print('rmse_train:', rmse_train, 'msle_train:', msle_train)
print('rmse_test:', rmse_test, 'msle_test:', msle_test)

# Count missing values in imputed training and testing data
print("Missing values in imputed training data:", np.isnan(train_X_imputed).sum())
print("Missing values in imputed testing data:", np.isnan(test_X_imputed).sum())
```

# Since we know the target isn't linear and has a lot of features, I used the HistGradientBoosting Regressor model which has native support for missing values (NaNs).

```python
from sklearn.experimental import enable_hist_gradient_boosting  # Enable the experimental HistGradientBoostingRegressor
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_squared_log_error
import pandas as pd
import numpy as np

# Assuming train_X, train_Y, test_X, test_Y are defined and contain your training and testing data

# Instantiate a HistGradientBoostingRegressor
model = HistGradientBoostingRegressor(random_state=42)

# Fit the model
model.fit(train_X, train_Y)

# Predict on training and testing data
train_pred_Y = model.predict(train_X)
test_pred_Y = model.predict(test_X)

# Clip negative predictions to 0
train_pred_Y = pd.Series(train_pred_Y.clip(0, train_pred_Y.max()), index=train_Y.index)
test_pred_Y = pd.Series(test_pred_Y.clip(0, test_pred_Y.max()), index=test_Y.index)

# Calculate RMSE and MSLE for both training and testing data
rmse_train = np.sqrt(mean_squared_error(train_pred_Y, train_Y))
msle_train = mean_squared_log_error(train_pred_Y, train_Y)
rmse_test = np.sqrt(mean_squared_error(test_pred_Y, test_Y))
msle_test = mean_squared_log_error(test_pred_Y, test_Y)

# Print evaluation metrics
print('rmse_train:', rmse_train, 'msle_train:', msle_train)
print('rmse_test:', rmse_test, 'msle_test:', msle_test)


```

# Visualisation of Results

```python
# To see how well the model works, a scatter plot of prediction against 
# ground truth is generated. The model gives a reasonable prediction 
# as the data points gather around the line y=x.
```

```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt

plt.figure(figsize=(18, 12))

# Concatenate train_Y and test_Y
train_test_Y = pd.concat([train_Y, test_Y])

# Concatenate train_pred_Y and test_pred_Y
train_test_pred_Y = pd.concat([train_pred_Y, test_pred_Y])

# Assuming 'data' is your DataFrame containing 'Country' and 'Region' columns
data_shuffled = data.loc[train_test_Y.index]
label = data_shuffled['Country']

colors = {'ASIA (EX. NEAR EAST)               ': 'red',
          'EASTERN EUROPE                     ': 'orange',
          'NORTHERN AFRICA                    ': 'gold',
          'OCEANA                             ': 'green',
          'WESTERN EUROPE                     ': 'blue',
          'SUB-SAHARAN AFRICA                 ': 'purple',
          'LATIN AMER. & CARIB                ': 'olive',
          'C.W. OF IND. STATES                ': 'cyan',
          'NEAR EAST                          ': 'hotpink',
          'NORTHERN AMERICA                   ': 'lightseagreen',
          'BALTICS                            ': 'rosybrown'}

for region, color in colors.items():
    X = train_test_Y.loc[data_shuffled['Region'] == region]
    Y = train_test_pred_Y.loc[data_shuffled['Region'] == region]
    plt.scatter(x=X, y=Y, marker='.', color=color, s=200, linewidths=0, label=region)

plt.legend(loc=4, prop={'size': 12})

plt.xlabel('GDP ($ per capita) ground truth', labelpad=40, fontsize=24)
plt.ylabel('GDP ($ per capita) predicted', labelpad=40, fontsize=24)

plt.plot([-1000, 50000], [-1000, 50000], c='gray')

plt.xlim(-1000, 60000)
plt.ylim(-1000, 40000)

for i in range(0, train_test_Y.shape[0]):
    if ((data_shuffled['Area (sq. mi.)'].iloc[i] > 8e5) or
            (data_shuffled['Population'].iloc[i] > 1e8) or
            (data_shuffled['GDP ($ per capita)'].iloc[i] > 10000)):
        plt.text(train_test_Y.iloc[i] + 200, train_test_pred_Y.iloc[i] - 200, label.iloc[i], size='small')

plt.show()
```
